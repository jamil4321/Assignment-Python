```python
print("===========Welcome to My TODO App===========")

flag = False
user_task = []
while flag!=True:
    print("""
    1. Insert Record
    2. Edit Record
    3. Delete Record
    4. Delete All
    5. Exit
    """)
    if len(user_task) > 0:
      print("======== Your Task ======")
    
      for i in range(len(user_task)) :
            print(i+1,".) ", user_task[i])
            
      print("======== End ===========")
    user_input = int(input("Enter the Task number do you Want to Perform : "))
    
    if user_input == 1:
        newTask = input("Insert New Task : ")
        user_task.insert(len(user_task),newTask)
    elif user_input == 2:
        if len(user_task) > 0:
            edit_task = int(input("Insert The Task Number do you want to edit"))
            print("Your Selected Task is: ", user_task[edit_task-1])
            promt = input("Do You Really Perform this opreation: ").lower()
            if promt == 'y' or promt == 'yes':
                user_task[edit_task-1] = input("Insert New Task : ")
    elif user_input == 3:
        if len(user_task) > 0:
            delete_record = int(input("Which Task You want to remove : "))
            print("Your Selected Task is: ", user_task[delete_record-1])
            promt = input("Do You Really Perform this opreation: ").lower()
            if promt == 'y' or promt == 'yes':
                del user_task[delete_record-1]
    elif user_input == 4:
        if len(user_task) > 0:
            promt = input("Do You Really Perform this opreation: ").lower()
            if promt == 'y' or promt == 'yes':
                user_task.clear()
    elif user_input == 5:
        promt = input("Do You Really Perform this opreation: ").lower()
        if promt == 'y' or promt == 'yes':
            flag = True
    

print("Thanks For Using My Todo App")
```
